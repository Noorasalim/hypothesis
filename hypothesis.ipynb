The company needs to check whether there is any dependency between thefeatures “Region” and “Manager”

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(data['Region'],data['Manager'])
```

```python
contingency_table
```

```python
from scipy.stats import ttest_rel
```

```python
stat, p, dof, expected = chi2_contingency(contingency_table)

print(f"The Test chi-square value is :\t{stat:.3f}")

chi2_critical = 9.488 # the chi2 value at alpha = 0.05 and dof = 4

if stat > chi2_critical:
    print(f" Null Hypothesis Rejected")
elif stat < chi2_critical:
    print ("Null Hypothesis Accepted")
```
